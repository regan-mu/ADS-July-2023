```python
import matplotlib.pyplot as plt
```

```python
# pyplot
# Object Oriented
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
x
```

```python
y
```

```python
#Pyplot
plt.plot(x, y)
plt.xlabel("Numbers")
plt.ylabel("Cosine of X")
plt.title("Numbers and their cosines")
plt.show()
```

```python
nums = list(range(1, 7))
sq = [n**2 for n in nums]
```

```python
plt.plot(nums, sq)
plt.xlabel("Numbers")
plt.ylabel("Squares of Numbers")
plt.title("Numbers and their squares")
plt.grid()
plt.show()
```

```python
sin = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y, color="red", label="Cosine")
ax.plot(x, sin, color="green", label="Sine")
ax.set_xlabel("Numbers", color="violet")
ax.set_ylabel("Cosine/Sine of X", color="violet")
ax.set_title("Numbers, their sines & cosines", color="magenta")
# ax.set_xticks(ticks=x, color="green")
plt.legend(loc="upper right")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 12, 8]

plt.bar(categories, values, color="magenta")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.savefig("bar.png")
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]

plt.pie(sizes, labels=labels, autopct="%.f%%")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(16, 7))

ax1.plot(x, y1)
ax1.set_title('sin(x)')
ax1.set_ylabel("Sin of X")

ax2.plot(x, y2)
ax2.set_title('cos(x)', color="cyan")
ax2.set_ylabel("Cosine of X")

plt.show()
```
